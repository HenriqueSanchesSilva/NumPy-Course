```python
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/HenriqueSanchesSilva/NumPy-Course/master/citrus.csv?token=GHSAT0AAAAAACNQFAGM6OM2ULFKROV2U676ZN4DDZQ'
```

```python
dados = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
```

```python
dados_t = dados.T
```

```python
diametro_laranja = dados_t[0, :5000]
diametro_toranja = dados_t[0, 5000:]
peso_laranja = dados_t[1, :5000]
peso_toranja = dados_t[1, :5000]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(diametro_laranja,peso_laranja)
plt.plot(diametro_toranja,peso_toranja)
plt.legend(['Laranjas','Toranjas'])
```

```python
XL = diametro_laranja
YL = peso_laranja
nL = len(XL)


cAngularL = (nL * np.sum(XL*YL) - np.sum(XL) * np.sum(YL))/ ( nL * np.sum(XL**2) - (np.sum(XL))**2 )
cLinearL = np.mean(YL) - cAngularL * np.mean(XL)
yL = cAngularL * XL + cLinearL
xL = diametro_laranja
```

```python
XT = diametro_toranja
YT = peso_toranja
nT = len(XT)


cAngularT = (nT * np.sum(XT*YT) - np.sum(XT) * np.sum(YT))/ ( nT * np.sum(XT**2) - (np.sum(XT))**2 )
cLinearT = np.mean(YT) - cAngularT * np.mean(XT)
yT = cAngularT * XT + cLinearT
xT = diametro_toranja
```

```python
plt.plot(xL,yL, color='red')
plt.plot(diametro_laranja,peso_laranja)
plt.plot(xT,yT, color='black')
plt.plot(diametro_toranja,peso_toranja)

plt.legend(['Reta Laranja','Laranjas','Reta Toranja','Toranjas'])
```

```python
np.random.seed(84)
coef_angulares = np.random.uniform(low=0,high=30,size=100)
normaL = np.array([])
for i in range(100):
    normaL = np.append(normaL, np.linalg.norm(diametro_laranja-(coef_angulares[i]*XL+17)))

dados = np.column_stack([normaL,coef_angulares])
np.savetxt('dadosLaranja.csv',dados,delimiter=',')
```

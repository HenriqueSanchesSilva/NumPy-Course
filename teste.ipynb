#Ao copiar arrays utilizando o = ele fica atrelado ao original porem se utilizar o np.copy o vinculo  quebrado

a = coeficiente angular                                                                        
n = numero de elementos                                                                        
y = que nesse caso eh moscow                                                                   
x = que nesse caso sao as datas


```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
(X**2).shape
```

```python
a = ( n * np.sum(X*Y) - np.sum(X) * np.sum(Y) ) / ( n * np.sum(X**2) - (np.sum(X))**2 )
```

Coeficiente Linear

Para calcular o coeficiente linear, usamos a seguinte frmula:



```python
b = np.mean(Y) - a * np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas,Moscow)
plt.plot(x,y)
```

com esta reta feita  possivel utilizala para localizar estimativas dentro do conjunto de dados quanto para tambem realizar projees futuras com a formula x*a+b, sendo x o ponto que voce deseja e a+b a soma dos coeficientes. exemplos abaixo:

```python
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5*a+b,'*r')
plt.plot(100,100*a+b,'*g')
plt.legend(['valores', 'linha y', 'ponto atual', 'ponto futuro',])
```
